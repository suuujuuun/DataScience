```python
def derivative_x(f, x, step_size):
    m = (f(x + step_size))
    return m

def my_function(x):
    return x**2

slope_at_2 = derivative_x(my_function,2,0.00001)

print(slope_at_2)
```

#Partial derivative

```python
from sympy.plotting import plot3d

x, y = symbols('x y')

f = 2*x**3 + 3*y**3

dx_f = diff(f,x)
dy_f = diff(f ,y)

print(dx_f)
print(dy_f)

plot3d(f)
```
